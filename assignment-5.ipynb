### 4. Determine the most frequently purchased item pairs by customers and display the top 20 pairs.

```python
from itertools import combinations

# ŸÑroup by Invoice and create a list of StockCodes for each invoice
invoice_items = random_sample.groupby('Invoice')['StockCode'].apply(list)

# Function that generates pairs of items from the list of items/produxts
def generate_pairs(item_list):
    return list(combinations(item_list, 2))

# call the function to generate pairs for each invoice, and then stack the pairs
pair_counts = invoice_items.apply(generate_pairs).explode().value_counts()

# print top 20 frequently purchased item pairs
top_pairs = pair_counts.head(20)
print("Top 20 frequently purchased item pairs are:")
print(top_pairs)
```

### 5. The retail store aims to introduce a new promotion targeting items with low sales volume. This promotion entails offering a discounted price for a pair of two different items, with one item being among the top 10 best sellers and the other from the bottom 10 sellers. Customers who purchase these two items together will qualify for the special pricing. Your task is to recommend 10 pairs of items to kickstart this campaign.


```python
# the top 10 least preferred products list
most_preferred_products = random_sample['Description'].value_counts()[:10]

# the bottom 10 least preferred products list
least_preferred_products = random_sample['Description'].value_counts().nsmallest(10)

# list of pairs
pair_list = [(most_preferred_products.index[i], least_preferred_products.index[i]) for i in range(min(len(most_preferred_products), len(least_preferred_products)))]

# print each pair
for pair in pair_list:
    print(pair[0], " , ", pair[1])
```
